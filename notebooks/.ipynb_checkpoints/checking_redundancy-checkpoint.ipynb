# Checking Redundancy

```python
import numpy as np
```

```python
s1 = np.random.rand(3,5)
s2 = s1
s2 = np.delete(s2, 0, axis=0)
s2 = np.append(s2, np.random.rand(1,5), axis=0)
print(s1)
print(s2)
```

```python
n = min(s1.shape)
s1_k = np.zeros([n] + list(s1.shape))
s1_k[0] = s1
for i in range(1,s1.shape[0]):
    s1_k[i,:-i,:] = s1[i:,:] 
print(s1_k)
```

```python
np.sum(s1_k == s2, axis=2)
```

```python
s1 = np.random.rand(5,5,3)
s2 = s1
s2 = np.delete(s2, 0, axis=2)
s2 = np.append(s2, np.random.rand(5,5,1), axis=2)
print(s1)
print(s2)
```

This is getting too complicated. Let's try a different approach.

## Specifying overlap shape

```python
s1 = np.random.rand(3,4,4)
s2 = s1
s2 = np.delete(s2, 0, axis=0)
s2 = np.append(s2, np.random.rand(1,4,4), axis=0)
print(s1)
print()
print(s2)
```

```python
overlap_shape = [2,4,4]
s3 = np.zeros([i - j for i,j in zip(s2.shape, overlap_shape)])
print(s3.shape)
```

```python
s3 = np.delete(s1, overlap_shape, axis=0)
print(s1)
print(s3.shape)
print(s3)
```
