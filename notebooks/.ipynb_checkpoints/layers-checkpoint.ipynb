## Vanilla tf version

```python
input_layer = tf.placeholder(tf.float32, shape=[3, 3])
noisy_input = input_layer + tf.random_normal(tf.shape(input_layer))
sess = tf.Session()
stats = []
for _ in range(1000):
    stats.append(sess.run(noisy_input, feed_dict={input_layer: np.zeros([3, 3])}))
stats = np.asarray(stats)
mean = np.mean(stats, axis=0)
stdev = np.std(stats, axis=0)
print("Mean:", mean)
print("Stdev:", stdev)
```

## Using noisy layer in layers.py

```python
phi = 4
channels= 3
num_outputs = 4
learning_rate = 0.00025
params_file_path = None
output_directory = "./noisy_input/results"
scope = ""

tf.reset_default_graph()
sess = tf.Session()
net_file = "./noisy_input/noisy_input.json"
network = create_network(net_file,
                         phi=phi, 
                         num_channels=channels, 
                         num_outputs=num_outputs,
                         learning_rate=learning_rate,
                         params_file=params_file_path,
                         output_directory=output_directory,
                         session=sess,
                         scope=scope)
```

```python
graph = tf.get_default_graph()
[print(n.name) for n in graph.as_graph_def().node]
```

```python
screen = graph.get_tensor_by_name("screen:0")
health = graph.get_tensor_by_name("health:0")
velocity = graph.get_tensor_by_name("velocity:0")
screen_noisy = graph.get_tensor_by_name("screen_noisy/add:0")
health_noisy = graph.get_tensor_by_name("health_noisy/add:0")
velocity_noisy = graph.get_tensor_by_name("velocity_noisy/add:0")

feed_dict={screen: np.zeros([1] + screen.get_shape().as_list()[1:]),
           health: np.zeros([1] + health.get_shape().as_list()[1:]),
           velocity: np.zeros([1] + velocity.get_shape().as_list()[1:])} 
print("Screen noisy:", sess.run(screen_noisy, feed_dict=feed_dict)[0, 0, :5, :5])
print("Health noisy:", sess.run(health_noisy, feed_dict=feed_dict))
print("Velocity noisy:", sess.run(velocity_noisy, feed_dict=feed_dict))
```

```python
s_stats, h_stats, v_stats = [], [], []
feed_dict={screen: np.zeros([1] + screen.get_shape().as_list()[1:]),
           health: np.zeros([1] + health.get_shape().as_list()[1:]),
           velocity: np.zeros([1] + velocity.get_shape().as_list()[1:])}
for _ in range(1000):
    s_stats.append(sess.run(screen_noisy, feed_dict=feed_dict))
    h_stats.append(sess.run(health_noisy, feed_dict=feed_dict))
    v_stats.append(sess.run(velocity_noisy, feed_dict=feed_dict))
stats = [np.asarray(s_stats), np.asarray(h_stats), np.asarray(v_stats)]
mean, stddev = [], []
for i in range(3):
    mean.append(np.mean(stats[i], axis=0))
    stddev.append(np.std(stats[i], axis=0))
    
print("Screen mean:", mean[0][0, 0, :5, :5])
print("Screen stddev:", stddev[0][0, 0, :5, :5])
print("Health mean:", mean[1])
print("Health stddev:", stddev[1])
print("Velocity mean:", mean[2])
print("Velocity stddev:", stddev[2])
```

```python
s_stats, h_stats, v_stats = [], [], []
feed_dict={screen: np.ones([1] + screen.get_shape().as_list()[1:]),
           health: 2 * np.ones([1] + health.get_shape().as_list()[1:]),
           velocity: 3 * np.ones([1] + velocity.get_shape().as_list()[1:])}
for _ in range(1000):
    s_stats.append(sess.run(screen_noisy, feed_dict=feed_dict))
    h_stats.append(sess.run(health_noisy, feed_dict=feed_dict))
    v_stats.append(sess.run(velocity_noisy, feed_dict=feed_dict))
stats = [np.asarray(s_stats), np.asarray(h_stats), np.asarray(v_stats)]
mean, stddev = [], []
for i in range(3):
    mean.append(np.mean(stats[i], axis=0))
    stddev.append(np.std(stats[i], axis=0))
    
print("Screen mean:", mean[0][0, 0, :5, :5])
print("Screen stddev:", stddev[0][0, 0, :5, :5])
print("Health mean:", mean[1])
print("Health stddev:", stddev[1])
print("Velocity mean:", mean[2])
print("Velocity stddev:", stddev[2])
```
