##  Noisy input
Adds noise to input layer(s).

### Local version

```python
input_layer = tf.placeholder(tf.float32, shape=[3, 3])
noisy_input = input_layer + tf.random_normal(tf.shape(input_layer))
sess = tf.Session()
stats = []
for _ in range(1000):
    stats.append(sess.run(noisy_input, feed_dict={input_layer: np.zeros([3, 3])}))
stats = np.asarray(stats)
mean = np.mean(stats, axis=0)
stdev = np.std(stats, axis=0)
print("Mean:", mean)
print("Stdev:", stdev)
```

## Dropout

### Local version

```python
tf.reset_default_graph()
x = tf.placeholder(tf.float32, shape=[3, 3], name="input")
is_training = tf.placeholder(tf.bool, name="is_training")
keep_prob = tf.placeholder(tf.float32, name="keep_prob")
input_shape = tf.shape(x)
rand = tf.random_uniform(input_shape)
mask = tf.cast(tf.less_equal(rand, keep_prob), tf.float32, name="mask")
out = tf.where(is_training,
               x * mask,
               x,
               name="out")
```

```python
graph = tf.get_default_graph()
for n in graph.as_graph_def().node: print(n.name) 
```

```python
graph = tf.get_default_graph()
sess = tf.Session()

for kp in [0.2, 0.5, 0.8]:
    print("Keep prob: %.1f" % kp)
    feed_dict = {x: np.ones([3,3]),
                 is_training: True,
                 keep_prob: kp}
    print("Training:")
    print(sess.run(out, feed_dict=feed_dict))

    feed_dict = {x: np.ones([3,3]),
                 is_training: False, 
                 keep_prob: kp}
    print("Testing:")
    print(sess.run(out, feed_dict=feed_dict))
    print()
```

```python
# Build larger input
tf.reset_default_graph()
x = tf.placeholder(tf.float32, shape=[100, 100], name="input")
is_training = tf.placeholder(tf.bool, name="is_training")
keep_prob = tf.placeholder(tf.float32, name="keep_prob")
input_shape = tf.shape(x)
rand = tf.random_uniform(input_shape)
mask = tf.cast(tf.less_equal(rand, keep_prob), tf.float32, name="mask")
out = tf.where(is_training,
               x * mask,
               x,
               name="out")

sess = tf.Session()
for kp in np.arange(0, 1.1, 0.1):
    print("Keep prob: %.1f" % kp)
    train_stats, test_stats = [], []
    for _ in range(1000):
        feed_dict = {x: np.ones([100,100]),
                     is_training: True,
                     keep_prob: kp}
        out_ = sess.run(out, feed_dict=feed_dict)
        train_stats.append(np.mean(out_))

        feed_dict = {x: np.ones([100,100]),
                     is_training: False, 
                     keep_prob: kp}
        out_ = sess.run(out, feed_dict=feed_dict)
        test_stats.append(np.mean(out_))

    print("Train mean:", np.mean(np.asarray(train_stats)))
    print("Test mean:", np.mean(np.asarray(test_stats)))
    print()
```

### layers.py version

#### keep_prob = 0.5, FC layer

```python
phi = 4
channels= 3
num_outputs = 4
learning_rate = 0.00025
params_file_path = None
output_directory = "./layers/dropout/results"
scope = ""

tf.reset_default_graph()
sess = tf.Session()
net_file = "./layers/dropout/dropout_fc.json"
network = create_network(net_file,
                         phi=phi, 
                         num_channels=channels, 
                         num_outputs=num_outputs,
                         learning_rate=learning_rate,
                         params_file=params_file_path,
                         output_directory=output_directory,
                         session=sess,
                         scope=scope)
```

```python
graph = tf.get_default_graph()
for n in graph.as_graph_def().node: print(n.name)
```

```python
screen = graph.get_tensor_by_name("screen:0")
is_training = graph.get_tensor_by_name("FC_1_DROPOUT/is_training:0")
fc1 = graph.get_tensor_by_name("FC_1/Leaky_Relu:0")
fc1_dropout = graph.get_tensor_by_name("FC_1_DROPOUT/Select:0")
input_ = np.random.rand(1, screen.get_shape().as_list()[1], screen.get_shape().as_list()[2], screen.get_shape().as_list()[3])
feed_dict={screen: input_,
           is_training: True} 

fc1_out = sess.run(fc1, feed_dict=feed_dict)
fc1_dropout_out = sess.run(fc1_dropout, feed_dict=feed_dict)
print("FC_1:", np.count_nonzero(fc1_out))
print(fc1_out[0, :20])
print("FC_1_DROPOUT:", np.count_nonzero(fc1_dropout_out))
print(fc1_dropout_out[0, :20])
```

```python
train_stats, test_stats = [], []
for _ in range(1000):
    feed_dict={screen: input_,
               is_training: True} 
    #fc1_out = sess.run(fc1, feed_dict=feed_dict)
    fc1_dropout_out = sess.run(fc1_dropout, feed_dict=feed_dict)
    train_stats.append(np.squeeze(fc1_dropout_out))
    
    feed_dict={screen: input_,
               is_training: False}
    #fc1_out = sess.run(fc1, feed_dict=feed_dict)
    fc1_dropout_out = sess.run(fc1_dropout, feed_dict=feed_dict)
    test_stats.append(np.squeeze(fc1_dropout_out))

train_stats = np.asarray(train_stats)
test_stats = np.asarray(test_stats)

print("Train mean:", np.mean(train_stats != 0))
print("Test mean:", np.mean(test_stats != 0))
```

```python
lines = plt.plot(np.mean(train_stats != 0, axis=0), label='train')
#lines = plt.plot(np.mean(train_stats != 0, axis=0), label='train') # all zero
```
